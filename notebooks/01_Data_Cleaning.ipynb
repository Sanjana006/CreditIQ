```python
# Data Manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Display Options
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../data/accepted_2007_to_2018Q4.csv"
)
```

```python
print(f"Rows : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
```

```python
df.info()
```

```python
df.describe().T
```

```python
selected_columns = [
    "loan_amnt",
    "term",
    "int_rate",
    "installment",
    "grade",
    "sub_grade",
    "emp_length",
    "home_ownership",
    "annual_inc",
    "verification_status",
    "purpose",
    "dti",
    "fico_range_low",
    "fico_range_high",
    "delinq_2yrs",
    "revol_util",
    "revol_bal",
    "open_acc",
    "total_acc",
    "pub_rec",
    "earliest_cr_line",
    "issue_d",
    "addr_state",
    "loan_status"
]

df = df[selected_columns]
```

```python
df.head()
```

```python
missing = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
      .to_frame("Missing Values")
)

missing["Missing %"] = (
    missing["Missing Values"] /
    len(df)
) * 100

missing
```

```python
missing = missing[missing["Missing Values"] > 0]

plt.figure(figsize=(10,5))

sns.barplot(
    data=missing,
    x=missing.index,
    y="Missing %",
    color="steelblue"
)

plt.xticks(rotation=90)
plt.title("Percentage of Missing Values")
plt.tight_layout()

plt.show()
```

```python
df["emp_length"] = df["emp_length"].fillna("Unknown")
```

```python
numeric_columns = [
    "annual_inc",
    "dti",
    "revol_util",
    "delinq_2yrs",
    "pub_rec"
]

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows :", duplicates)
```

```python
df = df.drop_duplicates()
```

```python
print(df.shape)
```

#### Convert Percentage Columns

```python
percentage_columns = [
    "int_rate",
    "revol_util"
]

for col in percentage_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace("%", "", regex=False)
    )

    df[col] = pd.to_numeric(df[col], errors="coerce")

df[["int_rate", "revol_util"]].head()
```

#### Convert Loan Term

```python
df["term"] = (
    df["term"]
    .astype(str)
    .str.extract(r"(\d+)", expand=False)
)

df["term"] = pd.to_numeric(df["term"], errors="coerce")
```

#### Convert Employment Length

```python
df["emp_length"] = (
    df["emp_length"]
    .replace({
        "< 1 year": 0,
        "10+ years": 10
    })
    .str.extract("(\d+)")
)

df["emp_length"] = pd.to_numeric(
    df["emp_length"],
    errors="coerce"
)
```

```python
df["emp_length"] = df["emp_length"].fillna(-1)
```

```python
df["emp_length"].isna().sum()
```

#### Convert Date Columns

```python
date_columns = [
    "issue_d",
    "earliest_cr_line"
]

for col in date_columns:
    df[col] = pd.to_datetime(
        df[col],
        format="%b-%Y",
        errors="coerce"
    )

df[date_columns].head()
```

#### Create FICO Score

```python
df["fico"] = (
    df["fico_range_low"] +
    df["fico_range_high"]
) / 2
```

```python
df.drop(
    columns=[
        "fico_range_low",
        "fico_range_high"
    ],
    inplace=True
)
```

#### Filter Loan Status

```python
valid_status = [
    "Fully Paid",
    "Charged Off",
    "Default"
]

df = df[
    df["loan_status"].isin(valid_status)
]
```

#### Check Outliers

```python
numeric_features = [
    "loan_amnt",
    "annual_inc",
    "dti",
    "revol_bal"
]

plt.figure(figsize=(12,8))

for i, col in enumerate(numeric_features):

    plt.subplot(2,2,i+1)

    sns.boxplot(
        y=df[col],
        color="skyblue"
    )

    plt.title(col)

plt.tight_layout()
plt.show()
```

Do we remove outliers?

No.

Extreme loan amounts, high incomes, or high DTI values are real borrowers, not data errors. Removing them would distort the portfolio and make the analysis less representative of actual lending operations.

Instead of deleting them, we'll retain these observations and use robust statistics (such as the median) where appropriate.
